```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Read the data from the CSV file

def graph(file, label):
    
    # Extract the 'N' and 'timing' columns
    N_values = file['N']
    timing_values = file['timing']

    # Plotting the data
    plt.plot(N_values, timing_values, marker='o', linestyle='-', label=label)

# Read data from three different files
file1 = pd.read_csv("mixed_method_time_vs_N.txt", delimiter=' ', header=None, names=['N', 'timing'])
file2 = pd.read_csv("chunk_method_time_vs_N.txt", delimiter=' ', header=None, names=['N', 'timing'])
file3 = pd.read_csv("extra_method_time_vs_N.txt", delimiter=' ', header=None, names=['N', 'timing'])

# Plot each file with a distinct label
graph(file1, label='Mixed Method')
graph(file2, label='Chunk Method')
graph(file3, label='Extra Method')

# Plotting settings
plt.title('N vs. Timing')
plt.xlabel('N values')
plt.ylabel('Timing values')
plt.xscale('log', base=2) 
plt.grid(True)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
